```python
# 1.Agregar producto
def add_product(name, price, quantity):
    for product in inventory:
        if product['name'] == name:
            product['quantity'] += quantity
            return
    inventory.append({"name" : name, "price" : price, "quantity" : quantity})
```

```python
# 2. ver inventario
def view_inventory():
    for product in inventory:
        print (f"Name: {product['name']}, Price: ${product['price']}, Quantity: {product['quantity']}")
```

```python
add_product("T-shirt", 50, 30)
add_product("pants", 25, 60)
add_product("bikini", 14, 10)
add_product("bikini", 15, 30)
```

```python
view_inventory()
```

```python
# 3. buscar producto
def search_product(name):
    for product in inventory:
        if product['name'] == name:
            print(f"Name: {product['name']}, Price: ${product['price']}, Quantity: {product['quantity']}")
            return
    print("Product not found")
```

```python
search_product("T-shirt")
```

```python
search_product("gloves")
```

```python
# 4. actualizar stock
def update_stock(name, quantity):
    for product in inventory:
        if product['name'] == name:
            product['quantity'] += quantity
            print("Stock updated")
            return
    print("Product not found") 
```

```python
# 5. eliminar el producto
def remove_product(name):
    for product in inventory:
        if product['name'] == name:
            inventory.remove(product)
            print("Product removed")
            return
    print("Product not found")
```

```python
remove_product("pants")
```

```python
# 6. calcular valor inventario
def calculate_inventory_value():
    total_sales = 0
    for product in inventory:
        total_sales += product['price'] * product['quantity']
    print(f"Total inventory value: ${total_sales}")
```

```python
# 7. realizar compra
def make_purchase():
    cart = []
    total_cost = 0

    view_inventory()

    while True:
        product_name = input("Enter product name (or 'exit' to finish): ")

        if product_name.lower() == 'exit':
            break

        quantity = int(input("Enter quantity: "))

        for product in inventory:
            if product['name'] == product_name and product['quantity'] >= quantity:
                product['quantity'] -= quantity
                cost = product['price'] * quantity
                total_cost += cost
                cart.append({'name': product_name, 'quantity': quantity, 'price': product['price']})
                break
        else:
            print("Product not available")

    print(f"Total purchase cost: ${total_cost}")  
    return cart, total_cost
```
